```python
from rolling.embedding import create_corpus_from_text, Corpus
file1:Corpus = create_corpus_from_text('texts/text1_atari200.txt', n=128)
file2:Corpus = create_corpus_from_text('texts/text2_shakespear.txt', n=128)
file3:Corpus = create_corpus_from_text('texts/text4_agent57.txt', n=128)

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
reduced_embeddings1 = PCA(n_components=2).fit_transform(file1.embeddings)
reduced_embeddings2 = PCA(n_components=2).fit_transform(file2.embeddings)
reduced_embeddings3 = PCA(n_components=2).fit_transform(file3.embeddings)

plt.figure()
plt.plot(reduced_embeddings1[:, 0], reduced_embeddings1[:, 1], color='blue', marker='o',   markersize=1, linewidth=0.1, label='Atari200')
plt.plot(reduced_embeddings2[:, 0], reduced_embeddings2[:, 1],  color='red', marker='o',   markersize=1, linewidth=0.1, label='Shakespear')
plt.plot(reduced_embeddings3[:, 0], reduced_embeddings3[:, 1],  color='green', marker='o', markersize=1, linewidth=0.1, label='Agent57')

plt.title('2D PCA of Two Embedding Series')
plt.legend()
plt.show()
```
